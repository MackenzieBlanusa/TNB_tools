```python
# make below cell into a function 
def open_ncfiles_in_directory(directory):
        """
        Docstring for open_ncfiles_in_directory
        
        :param directory: Description
        """
        varnames=[]
        for filename in os.listdir(directory):
            if filename.endswith(".nc"):
                file=os.path.splitext(os.path.basename(filename))[0]
                varnames.append(file)

        for name in varnames:
            globals()[name] = xr.open_dataset(directory+'/'+name+'.nc')
            print(name);

        print(f"\nLoaded {len(varnames)} NetCDF files")
```

```python
# Create function for renaming dimensions 

def rename_vars_if_exist(ds, rename_map):
    """
    Rename variables in an xarray Dataset only if they exist.

    Parameters
    ----------
    ds : xr.Dataset
        Input dataset
    rename_map : dict
        Mapping {old_name: new_name}

    Returns
    -------
    xr.Dataset
        Dataset with variables renamed where applicable
    """
    existing = {old: new for old, new in rename_map.items() if old in ds.data_vars}
    return ds.rename(existing)
```
